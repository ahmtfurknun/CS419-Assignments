```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def red_channel(p):
    result = np.empty_like(p)
    condition = p < 128
    result[condition] = 255 - 2 * p[condition]
    result[~condition] = 255
    return result

def green_channel(p):
    return p

def blue_channel(p):
    result = np.empty_like(p)
    condition = p < 128
    result[condition] = 255 - 2 * p[condition]
    result[~condition] = p[~condition] - 128
    return result


grayscale_values = np.arange(256)


# Red Channel
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(grayscale_values, red_channel(grayscale_values), label='Red Channel', color='red')
plt.xlabel('Grayscale Pixel Value')
plt.ylabel('Color Value')
plt.title('Red Channel Colorization Function')
plt.legend()

# Green Channel
plt.subplot(1, 3, 2)
plt.plot(grayscale_values, green_channel(grayscale_values), label='Green Channel', color='green')
plt.xlabel('Grayscale Pixel Value')
plt.ylabel('Color Value')
plt.title('Green Channel Colorization Function')
plt.legend()

# Blue Channel
plt.subplot(1, 3, 3)
plt.plot(grayscale_values, blue_channel(grayscale_values), label='Blue Channel', color='blue')
plt.xlabel('Grayscale Pixel Value')
plt.ylabel('Color Value')
plt.title('Blue Channel Colorization Function')
plt.legend()

plt.tight_layout()
plt.show()


image_path = "q1_kaplumbaga_terbiyecisi_osman_hamdi_bey.jpg"
grayscale_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Values in the range [0, 255]

colored_image = np.zeros((grayscale_image.shape[0], grayscale_image.shape[1], 3), dtype=np.uint8)
colored_image[:, :, 0] = red_channel(grayscale_image)
colored_image[:, :, 1] = green_channel(grayscale_image)
colored_image[:, :, 2] = blue_channel(grayscale_image)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(grayscale_image, cmap='gray')
plt.title('Input (Grayscale)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(colored_image)
plt.title('Output (Colorized)')
plt.axis('off')

plt.show()
```
